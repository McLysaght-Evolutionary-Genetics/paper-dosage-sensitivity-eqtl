```python
import pandas as pd
```

```python
tissues = ['Adipose',
'Blood',
'Embryo',
'Hypothalamus',
'Intramuscular_fat',
'Jejunum',
'Kidney',
'Leukocyte',
'Liver',
'Lung',
'Lymph_node',
'Macrophage',
'Mammary',
#'Mammary_L',
'Milk_cell',
'Monocytes',
'Muscle',
#'Muscle_cross',
#'Muscle_indicus',
#'Muscle_taurus',
'Ovary',
'Oviduct',
'Pituitary',
'Rumen',
'Spleen',
'Testis',
'Uterus']
```

```python
%%bash
for tissue_file in ../../datasets/cGTEx/cis-eQTL_Summary_Statistics/*.permutations.2rd.txt.gz; do
    tissue_name=$(basename $tissue_file .permutations.2rd.txt.gz)
    
    Rscript ../../datasets/cGTEx/scripts/4_cis-eQTL_p-value.nominal.correction_basePermutation.r ../../datasets/cGTEx/cis-eQTL_Summary_Statistics/${tissue_name}.permutations.2rd.txt.gz 0.05 ../../datasets/cGTEx/cis-eQTL_Summary_Statistics/${tissue_name}.nominals.2rd.txt.gz ../../datasets/cGTEx/cis-eQTL_Summary_Statistics/${tissue_name}.nominals.sig.txt
done
```

```python
%%bash
for tissue_file in ../../datasets/cGTEx/cis-eQTL_Summary_Statistics/*.permutations.2rd.txt.gz; do
    tissue_name=$(basename $tissue_file .permutations.2rd.txt.gz)
    
    Rscript ../../datasets/cGTEx/scripts/4_cis-eQTL_p-value.nominal.correction_basePermutation.MultipleTissueCorrection.r ../../datasets/cGTEx/cis-eQTL_Summary_Statistics/${tissue_name}.permutations.2rd.txt.gz 0.05 ../../datasets/cGTEx/cis-eQTL_Summary_Statistics/${tissue_name}.nominals.2rd.txt.gz ../../datasets/cGTEx/cis-eQTL_Summary_Statistics/significant_after_correction/${tissue_name}.nominals.sig.txt
done
```

```python
PCGenes = pd.read_csv('../../datasets/geneLists/Ensembl/CowV96/EnsV96CowProteinCodingGenes1-X.txt', sep='\t')
PCGenes.rename(columns={'Gene stable ID': 'Ensembl Gene ID'}, inplace=True)
PCGenes
```

```python
result = pd.DataFrame()
for tissue in tissues:
    tissueFrame = pd.read_csv('../../datasets/cGTEx/cis-eQTL_Summary_Statistics/'+tissue+'.permutations.2rd.txt.gz',
            sep='\s+',
            names=['Ensembl Gene ID',
                    'num_variants',
                    'MLE_shape1',
                    'MLE_shape2',
                    'dummy',
                    'best_variant',
                    'distance',
                    'nominal_pval',
                    'slope',
                    'corrected_pval_1',
                    'corrected_pval_2'])
    result = pd.concat([result, tissueFrame])
```

```python
geneTissueExpressionCounts = result['Ensembl Gene ID'].value_counts().to_frame()
geneTissueExpressionCounts.rename(columns={'Ensembl Gene ID':'expressedTissues'}, inplace=True)
geneTissueExpressionCounts['Ensembl Gene ID'] = geneTissueExpressionCounts.index
```

```python
geneTissueExpressionCounts
```

```python
result = pd.DataFrame()
for tissue in tissues:
    tissueFrame = pd.read_csv('../../datasets/cGTEx/cis-eQTL_Summary_Statistics/significant_after_correction/BonferroniCorrectedToo/'+tissue+'.nominals.sig.txt', sep='\t')
    tissueFrame = pd.DataFrame(tissueFrame['pid'].unique())
    result = pd.concat([result, tissueFrame])
```

```python
geneTissueeQTLCounts = result[0].value_counts().to_frame()
geneTissueeQTLCounts.rename(columns={0:'affectedTissues'}, inplace=True)
geneTissueeQTLCounts['Ensembl Gene ID'] = geneTissueeQTLCounts.index
```

```python
geneTissueeQTLCounts
```

```python
geneTissueExpressionCounts = pd.merge(geneTissueExpressionCounts, geneTissueeQTLCounts, on='Ensembl Gene ID', how='left').fillna(0)
geneTissueExpressionCounts
```

```python
geneTissueExpressionCounts = pd.merge(geneTissueExpressionCounts, PCGenes, on='Ensembl Gene ID')
len(geneTissueExpressionCounts)
```

```python
geneTissueExpressionCounts.to_csv('../../outputFiles/cGTExGenesWithCounts.csv', index=False)
```
