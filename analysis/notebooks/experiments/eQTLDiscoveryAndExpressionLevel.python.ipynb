# Prepare data for test of affect of gene expression and duplication status on eQTL discovery

```python
import pandas as pd
import mysql.connector
from sqlalchemy import create_engine
import re
import os.path
import numpy as np
```

```python
def removeGeneIDVersions(text):
    return re.findall('(ENSG\d+)', text)[0]
```

```python
medianExpressionValues = pd.read_csv('../../datasets/GTEx_Analysis_2016-01-15_v7_RNASeQCv1.1.8_gene_median_tpm.gct',
            sep='\t',
            skiprows=2)
medianExpressionValues = medianExpressionValues[medianExpressionValues.gene_id.str.contains("ENSGR") == False]
medianExpressionValues.rename(columns={'gene_id':'Ensembl Gene ID'}, inplace=True)
medianExpressionValues['Ensembl Gene ID'] = medianExpressionValues['Ensembl Gene ID'].apply(removeGeneIDVersions)
medianExpressionValues
```

```python
tissues = ['Adipose - Subcutaneous',
 'Adipose - Visceral (Omentum)',
 'Adrenal Gland',
 'Artery - Aorta',
 'Artery - Coronary',
 'Artery - Tibial',
 'Brain - Amygdala',
 'Brain - Anterior cingulate cortex (BA24)',
 'Brain - Caudate (basal ganglia)',
 'Brain - Cerebellar Hemisphere',
 'Brain - Cerebellum',
 'Brain - Cortex',
 'Brain - Frontal Cortex (BA9)',
 'Brain - Hippocampus',
 'Brain - Hypothalamus',
 'Brain - Nucleus accumbens (basal ganglia)',
 'Brain - Putamen (basal ganglia)',
 'Brain - Spinal cord (cervical c-1)',
 'Brain - Substantia nigra',
 'Breast - Mammary Tissue',
 'Cells - EBV-transformed lymphocytes',
 'Cells - Transformed fibroblasts',
 'Colon - Sigmoid',
 'Colon - Transverse',
 'Esophagus - Gastroesophageal Junction',
 'Esophagus - Mucosa',
 'Esophagus - Muscularis',
 'Heart - Atrial Appendage',
 'Heart - Left Ventricle',
 'Liver',
 'Lung',
 'Minor Salivary Gland',
 'Muscle - Skeletal',
 'Nerve - Tibial',
 'Ovary',
 'Pancreas',
 'Pituitary',
 'Prostate',
 'Skin - Not Sun Exposed (Suprapubic)',
 'Skin - Sun Exposed (Lower leg)',
 'Small Intestine - Terminal Ileum',
 'Spleen',
 'Stomach',
 'Testis',
 'Thyroid',
 'Uterus',
 'Vagina',
 'Whole Blood']
```

```python
affectedStatusWithExpressionLevel = pd.DataFrame()
for tissue in tissues:
   testedGenesInTissue = pd.read_csv('../../outputFiles/GTExV7/GTExTestedGenes/'+tissue+'.txt',
                sep='\t',
                names=['Ensembl Gene ID'])
    affectedGenesInTissue = pd.read_csv('../../outputFiles/GTExV7/GTExAffectedGenes/BonferroniCorrected/'+tissue+'.txt',
                sep='\t',
                names=['Ensembl Gene ID'])
    testedGenesInTissue.loc[testedGenesInTissue['Ensembl Gene ID'].isin(affectedGenesInTissue['Ensembl Gene ID']) , 'affected'] = 'Y'
    testedGenesInTissue.loc[~testedGenesInTissue['Ensembl Gene ID'].isin(affectedGenesInTissue['Ensembl Gene ID']) , 'affected'] = 'N'
    testedGenesInTissue = pd.merge(testedGenesInTissue, medianExpressionValues.loc[:,['Ensembl Gene ID',tissue]], on="Ensembl Gene ID")
    testedGenesInTissue = pd.merge(testedGenesInTissue, PCGenes, on="Ensembl Gene ID")
    testedGenesInTissue['tissue'] = tissue
    testedGenesInTissue.rename(columns={tissue:'medianExpression'}, inplace=True)
    affectedStatusWithExpressionLevel = pd.concat([affectedStatusWithExpressionLevel, testedGenesInTissue])
```
