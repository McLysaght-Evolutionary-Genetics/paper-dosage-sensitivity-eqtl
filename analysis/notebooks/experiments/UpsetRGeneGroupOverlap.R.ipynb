# Make plot of gene group overlap using UpSetR

```r
library(UpSetR)
library(ggplot2)
```

```r
?upset
```

```r
ohnologs <- read.csv("../../outputFiles/genesWitheQTLTissueCountBonferroniAndOhnologStatus.csv", sep="\t")
summary(ohnologs)
```

```r
CNVZarrei <- read.csv("../../outputFiles/genesWitheQTLTissueCountBonferroniAndCNVZarreiStatus.csv", sep="\t")
summary(CNVZarrei)
```

```r
CNVExAC <- read.csv("../../outputFiles/genesWitheQTLTissueCountBonferroniAndCNVExACStatus.csv", sep="\t")
summary(CNVExAC)
```

```r
haplo <- read.csv("../../outputFiles/genesWitheQTLTissueCountBonferroniAndHaploStatus.csv", sep="\t")
summary(haplo)
```

```r
CCN <- read.csv("../../outputFiles/genesWitheQTLTissueCountBonferroniAndCCNStatus.csv", sep="\t")
summary(CCN)
```

```r
CCN
```

```r
nrow(CCN$CCN == 'Y')
```

```r
upsetInput <- list(Ohnologs = ohnologs[ which(ohnologs$type=='ohno'), ]$Ensembl.Gene.ID,
                   SSDs = ohnologs[ which(ohnologs$type=='SSD'), ]$Ensembl.Gene.ID,
                   Singletons = ohnologs[ which(ohnologs$type=='singleton'), ]$Ensembl.Gene.ID,
                   `Within CNVR` = CNVZarrei[ which(CNVZarrei$CNV=='Y'), ]$Ensembl.Gene.ID,
                   `Outside CNVR` = CNVZarrei[ which(CNVZarrei$CNV=='N'), ]$Ensembl.Gene.ID,
                   `CNV-affected` = CNVExAC[ which(CNVExAC$CNV=='Y'), ]$Ensembl.Gene.ID,
                   `CNV-free` = CNVExAC[ which(CNVExAC$CNV=='N'), ]$Ensembl.Gene.ID,
                   `Haploinsufficient` = haplo[ which(haplo$haplo=='Y'), ]$Ensembl.Gene.ID,
                   `Haplosufficient` = haplo[ which(haplo$haplo=='N'), ]$Ensembl.Gene.ID,
                   `Not copy number conserved` = CCN[ which(CCN$CCN=='N'), ]$Ensembl.Gene.ID,
                   `Copy number conserved` = CCN[ which(CCN$CCN=='Y'), ]$Ensembl.Gene.ID)
```

```r
pdf(file="../../outputFiles/figures/UpsetRGeneOverlap.pdf",width = 12, height = 6) # or other device
upset_plot <- upset(fromList(upsetInput), order.by = "freq", nsets=length(upsetInput), nintersects=50)
upset_plot
dev.off()
```
